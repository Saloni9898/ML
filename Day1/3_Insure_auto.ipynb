```python
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('Insure_auto.csv')
```

```python
df
```

```python
X = df[['Home','Automobile']]
y = df['Operating_Cost']
df.dtypes
```

```python
poly = PolynomialFeatures(degree=3,include_bias=False)
poly
```

```python
X_poly = poly.fit_transform(X)
X_poly
```

```python
poly.get_feature_names_out()   # Home Automobile is interaction term 
```

```python
lr = LinearRegression()
lr.fit(X_poly,y)
```

```python
print('b0', lr.intercept_)
print('b1', lr.coef_)
```

# With Boston Dataset 

```python
boston = pd.read_csv('Boston.csv')
boston
```

```python
X = boston.drop('medv', axis=1)
y = boston['medv']
```

```python
poly = PolynomialFeatures(degree=2,include_bias=False)
```

```python
poly
```

```python
Boston_poly = poly.fit_transform(X)
```

```python
poly.get_feature_names_out()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(Boston_poly,y,test_size=0.3,random_state=24)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
print('b0', lr.intercept_)
print('b1', lr.coef_)
```

```python
x = lr.coef_
```

```python
len(x)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(r2_score(Y_test, y_pred))
```

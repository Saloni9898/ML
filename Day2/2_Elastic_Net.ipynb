```python
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import ElasticNet
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
boston = pd.read_csv('Boston.csv')
boston.head()
```

```python
a = [0.01,0.1,0.5,1,2.5, 3,5]
r = [0.001, 0.3, 0.5, 0.7 , 0.9]
```

```python
X = boston.drop('medv', axis=1)
y = boston['medv']
X_train, X_test, Y_train , Y_test = train_test_split(X,y,test_size=0.3, random_state=24)
```

```python
scores = []
for element in a:
    for l1 in r:
        el = ElasticNet(alpha = element , l1_ratio = l1)
        el.fit(X_train,Y_train)
        y_pred = el.predict(X_test)
        # print(r2_score(Y_test,y_pred))
        scores.append([element, l1 ,r2_score(Y_test,y_pred)])
        
df_scores = pd.DataFrame(scores, columns = ['Alpha', 'L1-ratio', 'r2-score'])
print('Best Parameters')
df_scores.sort_values('r2-score', ascending=False).iloc[0]
```

```python

print("scores : ", scores)
max = scores[0][2]

for i in scores:
    if max < i[2]:
        max = i[2]
        
print("Maximum r2_score is : ", max)      
print("Index of minimum r2_score : ", np.argmax(max))
len(scores)
```

```python
i_max = np.argmax(scores)
#print("Best Alpha : ", a[i_max])
```

```python
charges = []
for element in a:
    for l1 in r:
        el = ElasticNet(alpha = element , l1_ratio = l1)
        el.fit(X_train,Y_train)
        y_pred = el.predict(X_test)
        # print(r2_score(Y_test,y_pred))
        charges.append([element, l1 ,r2_score(Y_test,y_pred)])
        
df_charges = pd.DataFrame(charges, columns = ['Alpha', 'L1-ratio', 'r2-score'])
print('Best Parameters')
df_charges.sort_values('r2-score', ascending=False).iloc[0]
```

```python
# Chemical process file using mean imputer
df = pd.read_csv('ChemicalProcess.csv')
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
#impute = SimpleImputer(strategy='mean').set_output(transform= 'pandas')
impute = SimpleImputer(strategy='median').set_output(transform= 'pandas')
impute_data = impute.fit_transform(df)
df.isnull().sum()
```

```python
impute_data
```

```python
X = impute_data.drop('Yield',axis=1)
y = impute_data['Yield']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.3,random_state=24)
```

```python
a = [0.01,0.1,0.5,1,2.5, 3,5]
r = [0.001, 0.3, 0.5, 0.7 , 0.9]
```

```python
yields = []
for ele in a:
    for l1 in r:
        e1 = ElasticNet(alpha = ele, l1_ratio=l1)
        e1.fit(X_train,Y_train)
        y_pred = e1.predict(X_test)
        yields.append([ele, l1, r2_score(Y_test,y_pred)])
```

```python
yields
```

```python
df = pd.DataFrame(data = yields,columns=['Alpha','Element','r2_score'])
print('using Median Strategy')
df.sort_values('r2_score',ascending=False).iloc[0]
```

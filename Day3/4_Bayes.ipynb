```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import BernoulliNB
```

```python
df = pd.read_csv('tel_bayes.csv')
df

```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

```python
nb = BernoulliNB(alpha=0 , force_alpha=True)
X = df[['TT_gt_100_y', 'Gender_male']]
y= df['Response_not bought']
nb.fit(X,y)
```

```python
#dum_tst = np.array([['y', 'male']])
#dum_tst = np.array([[True, True]])   # TT>100 and male 
dum_tst = np.array([[False, False]])   #TT<100 and female
probs = nb.predict_proba(dum_tst)
print('P[Buying]', probs[0][0])
print('P[ Not Buying]', probs[0][1])
print(nb.predict(dum_tst))  #Gives outcome with highest probability
```

```python
#All possibbilities 

tst =  np.array([['y', 'male'], ['y', 'female'], ['n', 'male'], ['n', 'female']])  
tst = pd.DataFrame(tst, columns = ['TT_gt_100','Gender'])
dum_tst = pd.get_dummies(tst, drop_first=True)

nb.predict_proba(dum_tst)
```

```python
tel = pd.read_csv('Telecom.csv')
```

```python
tel
```

```python
X = tel.drop('Response',axis=1)
y = tel['Response']
```

```python
X_dum = pd.get_dummies(X,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_dum, y, test_size=0.2, random_state=42, stratify=tel['Response'])
```

```python

nb = BernoulliNB()
nb.fit(X_train,y_train)

y_pred = nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

# Cancer Dataset 

```python
can = pd.read_csv('Cancer.csv').set_index('subjid')
```

```python
can
```

```python
X = can.drop('Class', axis = 1)
y = can['Class']
```

```python
x_dummies = pd.get_dummies(X)
x_dummies
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_dummies, y, test_size=0.3, random_state=24)
```

```python
bb = BernoulliNB(alpha = 1, force_alpha=False)
bb.fit(X_train,y_train)
y_pred = bb.predict(X_test)
print(accuracy_score(y_test,y_pred))
```
